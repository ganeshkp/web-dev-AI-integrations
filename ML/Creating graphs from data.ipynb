```python
import matplotlib.pyplot as plt

#Effect of time spent walking (hours) on the distance travelled (miles)
time_spent_walking = [1,2,3,4,5]
distance = [2,4,6,8,10]

plt.xlabel('Time spent walking (hours)')
plt.ylabel('Distance travelled (miles)')
plt.title('Graph to show the effect of time spent walking (hours) on the distance travelled (miles)')
plt.plot(time_spent_walking, distance)
plt.show()
```

```python
#Effect of car age (years) on price (£)
car_age = [1,2,5,10,30]
price = [30000,25000,18000,10000,4000]

plt.xlabel('Car age (years)')
plt.ylabel('Price ($)')
plt.title('Graph to show the effect of car age (years) on price ($)')
plt.plot(car_age, price)
plt.show()
```
