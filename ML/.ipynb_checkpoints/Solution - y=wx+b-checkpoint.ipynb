```python
import matplotlib.pyplot as plt
#Effect of amount of water provided (litres) per day on the size of trees (m)

water =     [0,1,2,3,4,5,6,7,8]
tree_size = [4,5,6,7,8,9,10,11,12]


plt.xlabel('Water (litres)')
plt.ylabel('Size of tree (M)')
plt.title('Graph to show the effect of watering on tree size')
plt.plot(water, tree_size)
plt.show()
```

```python
tree_size[3] == (2*water[3]) + 4
```
